```python
cat=df.select_dtypes(include='object')
print('Categorical data ',list(cat))
print(len(list(cat)))
print()
num=df.select_dtypes(include=np.number)
print( 'numerical',list(num))
print(len(list(num)))
```

```python
for i in num:
    print(i)
    plt.subplot(1,2,1)
    sns.histplot(df[i],kde=True,element='step')
    plt.subplot(1,2,2)
    sns.boxplot(df[i])
    plt.show()
```

```python
for i in cat:
    print(i)
    sns.countplot(data=df,x=i)
    plt.show()
```

```python
df2 = df.select_dtypes(exclude=['object'])
sns.heatmap(df2.corr(),annot=True)
```
