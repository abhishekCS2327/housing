```python
df1=df.copy()
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_data=scaler.fit_transform(df1)
scaled_df=pd.DataFrame(scaled_data,columns=df1.columns)
scaled_df.head()
```

```python
df1
```

```python
x1=scaled_df.drop(columns='price')
```

```python
x2=scaled_df['price']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x1,x2,test_size=0.3,random_state=32)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(xtrain,ytrain)
model_prediction = model.predict(xtest)
```

```python
print(model.score(xtrain,ytrain)*100)
print(model.score(xtest,ytest)*100)
```
